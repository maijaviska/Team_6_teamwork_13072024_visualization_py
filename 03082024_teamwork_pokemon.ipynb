```python
# required libraries

import pandas as pd #analysis
import matplotlib.pyplot as plt #vizualizations
```

```python
pokemon_df = pd.read_csv('/content/Pokemon.csv')
```

```python
#1. How many Pokémons are with 'Type 1' == Water as a % of total?

water_type1 = pokemon_df[pokemon_df['Type 1']=='Water']
water_type_pct_of_total = round(len(water_type1)/len(pokemon_df)*100)
print(f'Water type Pokémons are {water_type_pct_of_total} % of all Pokémons')
```

```python
#2. What is the maximum 'Speed' value? What is the minimum 'Speed' value? What is the difference between max and min 'Speed'?

min_speed = pokemon_df['Speed'].min()
max_speed = pokemon_df['Speed'].max()
speed_diference = max_speed - min_speed
print(f'Maximum Pokémon speed is: {max_speed}\nMinimum Pokémon speed is: {min_speed}\nThe difference between max and min Pokémon speed is: {speed_diference}')
```

```python
#3. Filter the DataFrame to include only the Pokémon with 'Speed' >= 80.
#How many Pokémon meet this criterion? Display this DataFrame using your preferred visualization method.
pokemon_speed_above_80 = pokemon_df[pokemon_df['Speed']>=80]
count_above_80 = (len(pokemon_speed_above_80))
pokemon_speed_belove_80 = pokemon_df[pokemon_df['Speed']<80]
count_below_80 = (len(pokemon_speed_belove_80))

amounts = [count_above_80, count_below_80]
mylabels = ['Speed >= 80', 'Speed < 80']
mycolors = ['hotpink', 'gray']
myexplode = [0.1, 0]

plt.pie(amounts, labels = mylabels, colors = mycolors, explode = myexplode, shadow = True, autopct='%1.1f%%')
plt.title('Pokemon speed above and below 80')
```

```python
#4.(DIFFICULT) Find Pokémon with the longest name (excluding spaces)? What is this pokemons name?

pokemon_df['Name_length'] = pokemon_df['Name'].str.replace(' ','').str.len()

# Which pokemon has the longest name?

pokemon_df['Name_length'].max() # longest name without spaces
pokemon_df['Name_length'].idxmax() #index of the longest name

longest_pokemon_name = pokemon_df['Name'].iloc[pokemon_df['Name_length'].idxmax()]
print('Pokemon with the longest Name (excluding spaces) is:', longest_pokemon_name)
```
